```python
import pandas as pd
import numpy as np
from sqlite3 import connect
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
paises_migraciones=pd.read_csv(r'migration.csv')
paises_poblacion=pd.read_csv(r'population-and-demography.csv')
```

```python
conn = connect(':memory:') 
paises_migraciones.to_sql('pam',conn,index=False)
paises_poblacion.to_sql('pap',conn,index=False)
```

```python
paises_migraciones_america=pd.read_sql('SELECT Year,Country,Emigrants,`International migrants`, `Net migration`, `Net migration rate` FROM pam WHERE Country IN ("Argentina","Belice","Bolivia", "Brazil","Canada","Colombia", "Costa Rica", "Cuba", "Chile", "Ecuador", "El Salvador", "Guatemala", "Haiti", "Honduras", "Jamaica", "Mexico", "Nicaragua", "Panama", "Paraguay", "Peru","Puerto Rico", "Republica Dominicana","United States","Uruguay","Venezuela","Antigua y Barbuda","Bahamas","Barbados","Dominica","Granada","Guyana","Jamaica","República Dominicana","San Cristobal y Nieves","San Vicente y las Granadinas","Santa Lucia")',conn)
```

```python
paises_poblacion_america=pd.read_sql('SELECT `Country name`, Year, Population FROM pap WHERE `Country name` IN ("Argentina","Bolivia","Belice", "Brazil","Canada","Colombia", "Costa Rica", "Cuba", "Chile", "Ecuador", "El Salvador", "Guatemala", "Haiti", "Honduras", "Jamaica", "Mexico", "Nicaragua", "Panama", "Paraguay", "Peru","Puerto Rico", "Republica Dominicana","United States","Uruguay","Venezuela","Antigua y Barbuda","Bahamas","Barbados","Dominica","Granada","Guyana","Jamaica","República Dominicana","San Cristobal y Nieves","San Vicente y las Granadinas","Santa Lucia")',conn)
```

```python
tabla=pd.read_sql('SELECT a.Year,a.Country,a.Emigrants,a.`International migrants`,a.`Net migration`,a.`Net migration rate`,c.Population FROM a a JOIN c c ON(a.Year=c.Year AND a.`Country`=c.`Country name` )',conn)
```

```python
tabla.dropna(inplace=True)
tabla.reset_index(drop=True, inplace=True)
```

```python
tabla
```

```python
tabla['relative migration']=tabla['Emigrants']/tabla['Population']
```

```python
tabla['relative migration net']=tabla['Net migration']/tabla['Population']
```

```python
tabla['Inmigrants']=tabla['Net migration']+tabla['Emigrants']
```

```python
tabla
```

```python
tabla.sort_values('relative migration net')
```

```python
tabla.to_csv('tabla.csv')
```

```python
grouped_single = tabla.groupby('Country').agg({'relative migration net': ['mean', 'std', 'max']})
```

```python
grouped_single.count()
```

```python
grouped_single=grouped_single.sort_values(('relative migration net', 'mean'),ascending=True)
```

```python
grouped_single.to_csv('paises_prioridad.csv')
```

```python
grouped_single.sort_values(('relative migration', 'std'),ascending=False)
```
